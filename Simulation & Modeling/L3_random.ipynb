## Lab3: Generation of Random Number

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
a = 13  # multiplier
c = 37  # increment
m = 256  # mod
N = 200  # number of observations
k = 16  # number of cells
seed = 83  # start value
alpha = 0.99  # confidence level

# Generate pseudo-random sequence
x = np.zeros(N)
x[0] = seed
for i in range(1, N):
    x[i] = (x[i - 1] * a + c) % m

# Plot the histogram
plt.hist(x, bins=k, facecolor='g')
plt.grid()
plt.xlabel('Number')
plt.ylabel('Frequency')
plt.title('Histogram of pseudo-random sequence')
plt.show()
```

```python
# Perform chi-squared test
expected_freq = np.ones(k) * (N / k)
observed_freq, _ = np.histogram(x, bins=[i * m / k for i in range(k + 1)])
D = np.sum((observed_freq - expected_freq) ** 2 / expected_freq)
D
```
