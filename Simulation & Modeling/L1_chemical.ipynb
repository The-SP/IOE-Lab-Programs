## Lab 1: Simulation of the chemical reaction (continuous system)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k1 = 1
k2 = 2
n = 1000

c1 = np.zeros(n)
c2 = np.zeros(n)
c3 = np.zeros(n)

c1[0] = 1
c2[0] = 1
c3[0] = 0

dt=0.001
```

```python
for i in range(1, n):
    c1[i] = c1[i-1] + (k2 * c3[i-1] - k1 * c1[i-1] * c2[i-1]) * dt
    c2[i] = c2[i-1] + (k2 * c3[i-1] - k1 * c1[i-1] * c2[i-1]) * dt
    c3[i] = c3[i-1] + (2 * k1 * c1[i-1] * c2[i-1] - 2 * k2 * c3[i-1]) * dt
```

```python
t = np.arange(0, n*dt, dt)

# Create three subplots
plt.figure(figsize=(15, 5))
plt.suptitle('State of the system i.e. the value of reactants and product at different point of time')

# Plot for C1
plt.subplot(1, 3, 1)
plt.plot(t, c1, 'b')
plt.ylabel('C1')
plt.xlabel('t (sec)')

# Plot for C2
plt.subplot(1, 3, 2)
plt.plot(t, c2, 'r')
plt.ylabel('C2')
plt.xlabel('t (sec)')

# Plot for C3
plt.subplot(1, 3, 3)
plt.plot(t, c3, 'k')
plt.ylabel('C3')
plt.xlabel('t (sec)')

plt.tight_layout()
plt.show()
```
