## Lab 5: Simulation of National Econometric System

```python
import numpy as np
import matplotlib.pyplot as plt

G = np.array([20000, 30000, 25000, 35000, 28000, 32000, 27000, 40000, 29000, 38000, 26000, 31000, 33000, 27000, 36000, 31000])
Y = np.zeros(len(G))

I = np.zeros_like(G)
T = np.zeros_like(G)
C = np.zeros_like(G)

for i in range(1, len(G)):
    I[i] = 2 + 0.1 * Y[i - 1]
    Y[i] = 45.45 + 2.27 * (I[i] + G[i])
    T[i] = 0.2 * Y[i]
    C[i] = 20 + 0.7 * (Y[i] - T[i])

n = np.arange(len(G))
```

```python
plt.figure(figsize=(8, 6))
plt.grid(True)
plt.plot(n, Y, '-k', label='National Income', linewidth=2)
plt.plot(n, C, '--r', label='Consumption', linewidth=2)
plt.plot(n, G, ':b', label='Govt. Expenditure', linewidth=2)
plt.plot(n, T, '-.c', label='Tax', linewidth=2)
plt.plot(n, I, '-m', label='Investment', linewidth=2)

plt.title('National Econometric System')
plt.xlabel('Time (in years)')
plt.ylabel('Amount (in million dollars)')
plt.legend()
plt.show()
```
