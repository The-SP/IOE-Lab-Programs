```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# study = an array that contains the hours of study. This is the explanatory variable. <br>
# score = an array that contains the test scores. This is the response variable.


study = np.array([ 3, 4.5, 6, 1.2, 2, 6.9, 6.7, 5.5])
score = np.array([ 88, 85, 90, 80, 81, 92, 95, 90])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
study = study.reshape(-1, 1)
study.ndim
```

```python
# Create and fit the linear regression model
regression = LinearRegression()
regression.fit(study, score)

predicted_scores = regression.predict(study)
```

```python
# Plot the data and best fit line
plt.scatter(study, score, color='blue', label='Data')
plt.plot(study, predicted_scores, color='red', label='Best Fit Line')
plt.xlabel('Study')
plt.ylabel('Score')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
# Get the intercept, coefficient, total error
intercept = regression.intercept_
coefficient = regression.coef_[0]
error = np.sum((predicted_scores - score) ** 2)

print("Intercept:", intercept)
print("Coefficient:", coefficient)
print("Total Error:", error)
```
